```python
# find most popular course title
filt = df['num_subscribers'] == df['num_subscribers'].max()
df[filt]['course_title']
```

```python
# display 10 most popular courses as per number of subscribers
# mtd 1
top_10a = df.groupby('course_title')['num_subscribers'].mean().sort_values(ascending=False).head(10)
```

```python
sns.barplot( x='num_subscribers', y='course_title', data=top_10a)
```

```python
# mtd 2
top_10 = df.sort_values(by='num_subscribers', ascending=False)[['course_title', 'num_subscribers']].head(10)
```

```python
# lets visualise it
sns.barplot(y='num_subscribers', x='course_title', data=top_10)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.barplot( x='num_subscribers', y='course_title', data=top_10)
```

```python
# find the total number of courses related to python
filt_python = df['course_title'].str.contains('python', case=False)
len(df[filt_python])
```

```python
# display 10 most popular python courses as per number of subscribers
top_10_python = df[filt_python].sort_values(by='num_subscribers', ascending=False)[['course_title', 'num_subscribers']].head(10)

# lets see
sns.barplot(x='num_subscribers', y='course_title', data=top_10_python)
```
